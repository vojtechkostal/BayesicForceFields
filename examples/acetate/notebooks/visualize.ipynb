# Posterior Visualization

This notebook is intentionally plotting-only. It loads results from
stage `06-learn` and regenerates posterior plots without rerunning
training or MCMC.

```python
from pathlib import Path

from bff.bayes.results import PosteriorResults
from bff.plotting import plot_corner, plot_marginals

ROOT = Path('.').resolve()
LEARN_DIR = ROOT.parent / '06-learn'
SPECS = ROOT.parent / '03-sample-local' / 'specs.yaml'

results = PosteriorResults.load(
    posterior=LEARN_DIR / 'posterior.pt',
    priors=LEARN_DIR / 'priors.pt',
    specs=SPECS,
)
results.prepare_samples()
results.map_estimates
```

```python
plot_corner(results)
plot_marginals(results, SPECS)
```
